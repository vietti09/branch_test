```python
import pandas as pd
```

```python
# The path to our CSV file
cpi = "semi_clean_2yeardata.csv"

# Read our Kickstarter data into pandas
raw_cpi_df = pd.read_csv(cpi)
raw_cpi_df.head()
```

```python
raw_cpi_df.columns
```

```python
cpi_df = raw_cpi_df.loc[:,['Country Name', 'Indicator Name', 'Attribute', '2014', '2015','2016']]
# cpi_df
cpi_df["Indicator Name"].unique()
cpi_df = cpi_df.loc[cpi_df['Indicator Name'] == "Consumer Price Index, All items"]
cpi_df = cpi_df.loc[cpi_df['Attribute'] == "Value"]
final_cpi_df = cpi_df.dropna()
final_cpi_df
```

```python
final_cpi_df.to_excel('final_cpi_data.xls')
```

```python
happiness2015 = "happiness_2015.csv" 
happiness2015_df = pd.read_csv(happiness2015)
happiness2015_df.head()
```

```python
happy_2015 = happiness2015_df.loc[:,['Country', 'Happiness Rank', 'Happiness Score']]
happy_2015.head()
```

```python
cpi_happy_2015 = pd.DataFrame.merge(final_cpi_df, happy_2015, on='Country')
cpi_happy_2015
```
